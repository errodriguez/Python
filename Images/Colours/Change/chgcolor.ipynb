```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
input_path = "Signature_bw.png"  # Update this if your file has a different name
img_original = Image.open(input_path).convert("RGBA")

# Display the image
plt.figure(figsize=(5, 3))
plt.imshow(img_original)
plt.axis('off')
plt.title("Original Signature (Black)")
plt.show()
```

```python
# Make a copy to modify
img_blue = img_original.copy()
pixels = img_blue.load()

# Define your "blue ink" color (customize as desired)
blue_ink = (10, 60, 200)  # A nice inky blue

# Replace dark pixels with blue
for y in range(img_blue.height):
    for x in range(img_blue.width):
        r, g, b, a = pixels[x, y]
        if a > 0 and r < 50 and g < 50 and b < 50:  # Dark pixel detection
            pixels[x, y] = (*blue_ink, a)

# Display the result
plt.figure(figsize=(5, 3))
plt.imshow(img_blue)
plt.axis('off')
plt.title("Recolored Signature (Blue Ink)")
plt.show()

# Optional: save the result
img_blue.save("signature_blue.png")
```
